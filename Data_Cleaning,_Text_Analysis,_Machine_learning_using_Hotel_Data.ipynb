```python
import pandas as pd #Used for high level Data Indexing processes
import numpy as np  #Numpy provides fast mathematical function processing
import sklearn #A overall Machine Learning library for Data Analysis called SCIKIT LEARN for many other machine learning libraries 
```

```python
Hoteldata = pd.read_csv('https://raw.githubusercontent.com/Worlddatascience/Data_Sets/main/hotelReviewsInParis__en2019100120191005.csv')

```

```python
Hoteldata
```

```python
#I want to get a clearer picture of the data
#Perform some basic cleaning of the data by removing irrelevant characters

#make all data lower case
Hoteldata['review_body'] = Hoteldata['review_body'].str.lower()

#Removes non text data characters
Hoteldata['review_body'] = Hoteldata['review_body'].str.replace(r'|\!|\?|\'|,|-|\(|\)',"",)

#Fill null or nan values with ''
Hoteldata['review_body'] = Hoteldata['review_body'].fillna('')
```

```python
#taking a look at my data table once again
#the first 10 rows
#.head function gives us a sample of our data frame and the number within the () gives us that number
#Python starts with number 0
Hoteldata.head(10)
```

```python
#Remove unwanted columns
#irplace=true tells the data frame to make permanent changes
Hoteldata.drop(['Unnamed: 0','hotelUrl'], axis=1, inplace=True)
```

```python
Hoteldata.head(10)
```

commonly used in Natural language processing(a form of Reinforcement learning)

**Reinforcement learning** is a form of Machine learning

```python
#Import and initiate a count vectorizer
#A count vectorizer is used to train my model to recogize vocabulary
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Next I create max features to set a list on how many words to create columns for 
#The norm is 1000, 5000, 10000, and 50000
#THe more words in the document, the more max features
#For example, if this was a legal document with millions of rows my max features would be 50000 or more
vectorizer = CountVectorizer(max_features=5000)
```

```python
# then Vectorize review data to transform sentences into columns
X = vectorizer.fit_transform(Hoteldata['review_body'])
```

```python
#Every word in a fixed vocabulary is represented by a vector
#Vectorizing is the process of encoding text as integers i.e. numeric form to create feature vectors so that machine learning can understand our data
#Bag of words (BoW) or CountVectorizer describes the presence of frequency of words (this is how it learns)
#It can then be used later in another model to predict the proper context of words
bag_of_words = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
```

```python
#I have renamed some columns for easier understanding to target audience
Hoteldata.rename(columns={'review_body': 'hotel_reviews', 'hotelName': 'Hotel_Name'}, inplace=True)
```

```python
Hoteldata.head(10)
```

```python
#This code will give us the different Hotel names and how many times they are mentioned 
Hoteldata['Hotel_Name'].value_counts()
```

```python
#I then join (bag of words) with initial structured hotel data
Hoteldatafinal = Hoteldata.join(bag_of_words)
```

```python
#X will equal my bag of words
#This will help me to get accurate results in the end
X = bag_of_words
```

```python
#Hotel name (the outcome I care about) based on words in the review
Final_Hotel_Name = Hoteldata['Hotel_Name']
```

```python
#Import a random forest model.
#Random forest is a type of Supervised Learning Algorithm for Classification and Regression purposes
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()

#Fit that random forest model to our data.
rfc.fit(X, Final_Hotel_Name)
```

```python
Hotel_Review = ['clean','pets','food','pool']
```

```python
#COnvert my hotel reiew into a vector because of its flexibility with new data
#converting it into a vector makes my model more accurate
X_test = vectorizer.transform(Hotel_Review).toarray()
```

```python
#Random Forest Classifer
#Now we are running our last data set through our machine learning model
#Amongst the best Algorithm to use (seen as a better overall Algorithm for text/categorical data)
#It gives a result of 1(unless its more than 1) if present in the sentence and 0 if not present

prediction = rfc.predict(X_test)[3]
```
