```python
import pandas as pd #Used for high level Data Indexing processes
import numpy as np  #Numpy provides fast mathematical function processing
# Start with loading all necessary libraries
from PIL import Image
#Pillow is a very good Python Image Library
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
#I am importing necessary packages for word cloud
import matplotlib.pyplot as plt #high level library for data visualations 
%matplotlib inline
import urllib #data scraping/// allows python to access websites (library)
import requests #send HTTP requests using Python and connect with APIs
```

```python
#Second, I manually add words from Resume
#Add 1 word 10 times, another word 9 times, another word 8 times, another word 7 times, until you add words to get to 1
text=("Dadrian Dadrian Dadrian Dadrian Dadrian Dadrian Dadrian Dadrian Dadrian Dadrian  Python Python Python Python Python Python Python Python Python Madison Madison Madison Madison Madison Madison Madison Madison Daisy Daisy Daisy Daisy Daisy Daisy Daisy Donnie Donnie Donnie Donnie Donnie Donnie Fat Fat Fat Fat Fat Mama Mama Mama Mama Girls Girls Girls Church Church Sunday")

#Third I create the wordcloud object & shape
#capitalize WordCloud
wordcloud = WordCloud(width=512, height=480, margin=0).generate(text)

#fourth I insert the code to display the generated image:
plt.imshow(wordcloud, interpolation='bilinear') #interpolation creates a 2d visual 
plt.axis("off")
plt.margins(x=30, y=0)
plt.show()
```

```python
#This is the code to make wordcloud into shapes
#Website is clker.com/cliparts to change shapes
mask = np.array(Image.open(requests.get('http://www.clker.com/cliparts/O/i/x/Y/q/P/yellow-house-hi.png', stream=True).raw))
```

```python
def generate_wordcloud(text, mask):
    word_cloud = WordCloud(width= 512, height= 512, background_color='white', stopwords=STOPWORDS, mask=mask).generate(text)
    plt.figure(figsize=(10,8),facecolor= 'Gray', edgecolor='green')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

```python
#The more words the better, this wil take the shape of the clipart (in this example a house)
generate_wordcloud(text, mask)
```

```python
#The data set we will be using today is "The top 1000 baby names from 1880-2009"
Babynames=pd.read_csv("https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv")
#The above function is how you import a csv file, located in a github repository.
#Depending on the file and the file location, there is a workaround for importing files
#The simplest way to me is uploading dataset you will be working with directly into Github and import it 
```

```python
#I want to findout the most poplular baby name from 1880-2009
#Now we do some data transforamtion and clean the dataframe for our answer
#I will create a new data table just the colums for my wordcloud called "Babynames2"
Babynames2=Babynames ['name']
```

```python
#Check my data set
Babynames2.head(7)
```

```python
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(
    background_color='white',
    stopwords=STOPWORDS,
    max_words=200,
    max_font_size=100,
    random_state=42
    ).generate(str(Babynames2))

#Dispaly the generated image:
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()    
```

```python
#I notice the column name of my data frame is showing in my wordcloud
#I also realize object, length, and dtype also shouldnt be there
#So now I want to remove it
#I now to remove object, length, and dtype from the wordcloud by using the STOPWORDS function
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(
    background_color='white',
    stopwords=STOPWORDS,
    max_words=200,
    max_font_size=80,
    random_state=0
    ).generate(str(Babynames2))

#Dispaly the generated image:
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
